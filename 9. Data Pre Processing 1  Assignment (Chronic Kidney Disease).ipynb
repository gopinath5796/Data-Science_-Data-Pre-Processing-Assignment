```python
# Importing the pandas and it is named as pd.
import pandas as pd
```

```python
#'pd.read_csv' is used to read the "kidney_disease.csv" and it is stored in the 'dataset' Variable name.
dataset=pd.read_csv("kidney_disease.csv")
```

```python
# Calling the 'dataset' variable name to view the csv file stored into it.
dataset
```

```python
# Using the "isna().Sum()" command for showing the dataset has how many null Values in the dateaset.
dataset.isna().sum()
```

```python
# Using "Dataset.info()" can see the datatypes present in the dataset.
dataset.info()
```

```python
#Here using the "if-else" condition is used for seperating the datatypes from the dataset like (int,float) which is stored in Quan Variable name in the  and (object) which is stored in Qual Variable name by using the list.
#Quan contains only the Numerical values columns names and Quan contains only the Categorical values columns names both are stored in the QuanQual Variable Name.
def QuanQual(dataset):
    Quan=[]
    Qual=[]
    for columnName in dataset.columns:
        print(columnName)
        if(dataset[columnName].dtypes=="O"):
            #print("Qual")
            Qual.append(columnName)
        else:
            #print("Quan")
            Quan.append(columnName)
    return Quan,Qual
```

```python
# Both quan and qual in the same dataset in the name of QuanQual. Here we are seperating them as quan and qual.
quan,qual=QuanQual(dataset)
```

```python
# Here calling the quan for viewing the "int" column names stored in "quan" variable name
quan
```

```python
#Here we are calling the "quan" column names from the dataset.
dataset[quan]
```

```python
# Using the "isna.sum" for viewing the null values in the "quan dataset".
dataset[quan].isna().sum()
```

```python
# Here calling the quan for viewing the "object" column names stored in "qual" variable name
qual
```

```python
#Here we are calling the "qual" column names from the dataset.
dataset[qual]
```

```python
# Using the "isna.sum" for viewing the how many null values in the "qual dataset".
dataset[qual].isna().sum()
```

```python
# SimpleImputer which is used to 'replace' the 'multiple columns' null values in the dataset.
# Here we are using the 'mean' value for replacing the null values in the "quan dataset". We can use mean, median or mode (most_frequent) also.
# The mean values replaced in the dataset and and it forms like an array and the whole values are stored in the "quandataset" variable name.
# Here why we are using the 'mean' means all the values stored in the dataset was not having any outliers and also the mode also not suitable for this dataset.
import numpy as np
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan,strategy='mean')
imp.fit(dataset[quan])
quandataset=imp.transform(dataset[quan])
```

```python
imp
```

```python
# Here we are calling the "quandataset" to view the replaced values and all the values stored like an array.
quandataset
```

```python
# Here we are creating the table in the "quan" column names and passing the "quandataset" array in to the table. 
#Then the table is stored in the variable name of "quandataset"
quandataset=pd.DataFrame(quandataset,columns=['id','age', 'bp', 'sg', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo'])
```

```python
# Here we are calling the "quandataset" to view the table.
quandataset
```

```python
# Using the "isna.sum" for viewing the how many null values in the "quandataset". 
# Here we are verifing that onceagain still there is any null values in the dataset.
# Finally,all the "int datatype" columns are no null values in the "quandataset".
quandataset.isna().sum()
```

```python
# SimpleImputer which is used to 'replace' the 'multiple columns' null values in the dataset.
# Here we are using the 'mode (or) most_frequent' value for replacing the null values in the "qual dataset". 
# The 'mode (or) most_frequent' values replaced in the dataset and and it forms like an array and the whole values are stored in the "qualdataset" variable name.
# Here why we are using the 'mode (or) most_frequent' means the categorical data can be replaced only by using this "most_frequent" function. There is no otherway to use the mean or median here.
import numpy as np
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
imp.fit(dataset[qual])
qualdataset=imp.transform(dataset[qual])
```

```python
imp
```

```python
# Here we are calling the "qualdataset" to view the replaced values and all the values stored like an array.
qualdataset
```

```python
# Here we are creating the table in the "qual" column names and passing the "qualdataset" array in to the table. 
#Then the table is stored in the variable name of "qualdataset"
qualdataset=pd.DataFrame(qualdataset,columns=['rbc','pc','pcc','ba','pcv','wc','rc','htn','dm','cad','appet','pe','ane','classification'])
```

```python
# Here we are calling the "qualdataset" to view the table.
qualdataset
```

```python
# Using the "isna.sum" for viewing the how many null values in the "qualdataset". 
# Here we are verifing that onceagain still there is any null values in the dataset.
# Finally,all the "object datatype" columns are no null values in the "qualdataset".
qualdataset.isna().sum()
```

```python
#Here using the "join" command we are joing the quandataset and qualdataset.
#Both the joined dataset will be stored in the Variable Name of "finaldataset"
finaldataset = quandataset.join(qualdataset)
```

```python
# Here we are calling the "finaldataset" variable name for verifying that both the dataset are joined properly.
finaldataset
```

```python
# Finally we are verifying there is any null values in the "finaldataset" by using "isna().sum()" command.
finaldataset.isna().sum()
```

```python
# Finally we are verifying the dataset, whether all the rows and columns are there and its datatyper are correct by using the ".info()"Command.
finaldataset.info()
```

```python
# Finally we are converting the 'finaldateset' into 'csv file format' by giving the name of 'CKD_Pre-processed_Dataset.csv'.
# (header=True) means it save including the columnname, If False means it will delete the columnname.
# (index=False) means it will not create the sl.no. column. If True means it will create the new si.no column. 
# (Already the dataset contains the unique id like si.no., so only index=False given).
finaldataset.to_csv('CKD_Pre-processed_Dataset.csv', header=True, index=False)
```
